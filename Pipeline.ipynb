```python
import pandas as pd
```

```python
df=pd.read_excel("cleaned_accident_data.xlsx")
df
```

```python
# Drop redundant or highly correlated features
df.drop(['Urban_or_Rural_Area', 'Local_Authority_District'], axis=1, inplace=True)
```

```python
df.columns
```

```python
from sklearn.preprocessing import LabelEncoder
X=df.iloc[:,:-1]
y=LabelEncoder().fit_transform(df['Accident_Severity'])
```

```python
X
```

```python
y
```

```python
from sklearn_pandas import DataFrameMapper
from sklearn.preprocessing import StandardScaler,LabelEncoder
from xgboost import XGBClassifier
```

```python
dm=DataFrameMapper([
    (['Accident_months'],StandardScaler()),
    (['Day_of_Week'],LabelEncoder()),
    (['Junction_Control'],LabelEncoder()),
    (['Junction_Detail'],LabelEncoder()),
    (['Light_Conditions'],LabelEncoder()),
    (['Carriageway_Hazards'],LabelEncoder()),
    (['Number_of_Casualties'],StandardScaler()),
    (['Number_of_Vehicles'],StandardScaler()),
    (['Police_Force'],LabelEncoder()),
    (['Road_Surface_Conditions'],LabelEncoder()),
    (['Road_Type'],LabelEncoder()),
    (['Speed_limit'],LabelEncoder()),
    (['Weather_Conditions'],LabelEncoder()),
    (['Vehicle_Type'],LabelEncoder())
    
])
```

```python
dm
```

```python
ml=[]
ml.append(['mapper',dm])
# ml.append(['scaler',StandardScaler])
ml.append(['xgb',XGBClassifier(eval_metric='mlogloss', random_state=42)])
```

```python
ml
```

```python
from sklearn.pipeline import Pipeline
import warnings
warnings.filterwarnings('ignore')
```

```python
model=Pipeline(ml)
model.fit(X,y)
```
